```python
import numpy as np
from array import *
import matplotlib.pyplot as plt
import math
```

```python
#Read CVN Score for each lorentz factor and atmospheric neutrinos

CVN_b1p1_m05 = np.loadtxt('CVN_bdm_m05_b1p1_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
CVN_b1p5_m05 = np.loadtxt('CVN_bdm_m05_b1p5_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
CVN_b10_m05 = np.loadtxt('CVN_bdm_m05_b10_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)

CVN_b1p1_m10 = np.loadtxt('CVN_bdm_m10_b1p1_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
CVN_b1p5_m10 = np.loadtxt('CVN_bdm_m10_b1p5_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
CVN_b10_m10 = np.loadtxt('CVN_bdm_m10_b10_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)

CVN_b1p1_m20 = np.loadtxt('CVN_bdm_m20_b1p1_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
CVN_b1p5_m20 = np.loadtxt('CVN_bdm_m20_b1p5_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
CVN_b10_m20 = np.loadtxt('CVN_bdm_m20_b10_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)

CVN_b1p1_m40 = np.loadtxt('CVN_bdm_m40_b1p1_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
CVN_b1p5_m40 = np.loadtxt('CVN_bdm_m40_b1p5_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
CVN_b10_m40 = np.loadtxt('CVN_bdm_m40_b10_00a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)

CVN_Atm = np.loadtxt('CVN_Atm.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
```

```python
plt.rc('text', usetex=True)
plt.rc('font', size=12)
bestCutRight = 0.99
plt.figure(dpi=150)
h1, x1, _ = plt.hist(CVN_Atm[CVN_Atm>-1], bins= 40, density=True, label=r'Atm-$\nu$', histtype='step', lw =1)
h2, x2, _ = plt.hist(CVN_b1p1_m05[CVN_b1p1_m05>-1], bins= 40, density=True, label=r'$m_\chi=5 \textrm{ GeV}$', histtype='step', lw =1)
h3, x3, _ = plt.hist(CVN_b1p1_m10[CVN_b1p1_m10>-1], bins= 40, density=True, label=r'$m_\chi=10 \textrm{ GeV}$', histtype='step',lw =1)
h4, x4, _ = plt.hist(CVN_b1p1_m20[CVN_b1p1_m20>-1], bins= 40, density=True, label=r'$m_\chi=20 \textrm{ GeV}$', histtype='step',lw =1)
h5, x5, _ = plt.hist(CVN_b1p1_m40[CVN_b1p1_m40>-1], bins= 40, density=True, label=r'$m_\chi=40 \textrm{ GeV}$', histtype='step',lw =1)
h = max(h1.max(),h2.max(),h3.max(), h4.max(), h5.max())
plt.xlabel('CVN Score')
plt.ylabel('Normalized Frequency')
plt.vlines(bestCutRight, 0, h.max(), colors='#FFA500', linestyles='solid')
plt.plot(bestCutRight+0.03, h.max()/2, marker=r'$\rightarrow$', markersize=20,markerfacecolor='#FFA500',markeredgecolor="orange")
plt.legend(title = r'$\gamma = 1.1,$ hA+BR',loc='upper center')
plt.show()
```
