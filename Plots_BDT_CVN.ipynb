```python
#Read BDT Score for each lorentz factor and atmospheric neutrinos

BDT_b1p1_m05 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m05_b1p1.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
BDT_b1p5_m05 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m05_b1p5.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
BDT_b10_m05 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m05_b10.txt', delimiter=' ', skiprows=1,  dtype=np.float32)

BDT_b1p1_m10 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m10_b1p1.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
BDT_b1p5_m10 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m10_b1p5.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
BDT_b10_m10 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m10_b10.txt', delimiter=' ', skiprows=1,  dtype=np.float32)

BDT_b1p1_m20 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m20_b1p1.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
BDT_b1p5_m20 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m20_b1p5.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
BDT_b10_m20 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m20_b10.txt', delimiter=' ', skiprows=1,  dtype=np.float32)

BDT_b1p1_m40 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m40_b1p1.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
BDT_b1p5_m40 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m40_b1p5.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
BDT_b10_m40 = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_m40_b10.txt', delimiter=' ', skiprows=1,  dtype=np.float32)

BDT_Atm = np.loadtxt('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_Atm_02a.txt', delimiter=' ', skiprows=1,  dtype=np.float32)
```

```python
plt.rc('text', usetex=True)
plt.rc('font', size=12)
bestCutRight = 0.17
plt.figure(dpi=150,figsize=(6,4))
h1, x1, _ = plt.hist(BDT_Atm, bins= 30, density=True, label=r'Atm-$\nu$', histtype='step', lw =2)
h2, x2, _ = plt.hist(BDT_b1p1_m05, bins= 30, density=True, label=r'$m_\chi=5 \textrm{ GeV}$', histtype='step',lw =2)
h3, x3, _ = plt.hist(BDT_b1p1_m10, bins= 30, density=True, label=r'$m_\chi=10 \textrm{ GeV}$', histtype='step',lw =2)
h4, x4, _ = plt.hist(BDT_b1p1_m20, bins= 30, density=True, label=r'$m_\chi=20 \textrm{ GeV}$', histtype='step',lw =2)
h5, x5, _ = plt.hist(BDT_b1p1_m40, bins= 30, density=True, label=r'$m_\chi=40 \textrm{ GeV}$', histtype='step',lw =2)
h = max(h1.max(),h2.max(),h3.max(), h4.max(), h5.max())
plt.xlabel('BDT Score')
plt.ylabel('Normalized Frequency')
plt.vlines(bestCutRight, 0, h.max(), colors='#FFA500', linestyles='solid')
plt.plot(bestCutRight+0.02, h.max()/2, marker=r'$\rightarrow$', markersize=20,markerfacecolor='#FFA500',markeredgecolor="orange")
plt.legend(title = r'$\gamma = 1.1,$ hA+ESF',loc='upper left')
plt.savefig('FinalPlots/OptimalCuts/NuclearModelsOptSelectionApplication_MaxCVN/hA_ESF/BDT_b1p1'+str(time.strftime("%Y%m%d"))+'.pdf', format='pdf')
plt.show()
```
